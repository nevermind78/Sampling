# Sampling ( Ligne )
## Générer des échantillons aléatoires de taille donné ( nombre de ligne) à partir d'un dataset avec `pandas`

## Générer des échantillons aléatoires d'une seule ligne

```python
# importing pandas package 
import pandas as pd 

# making data frame from csv file 
data = pd.read_csv("Datasets/employees.csv") 

# generating one row 
row1 = data.sample(n = 1) 

# display 
row1 


```

```python
# generating another row 
row2 = data.sample(n = 1) 

# display 
row2 
```

### Générer un échantillon de taille n quelconque

```python
# generating n=25 rows
row2 = data.sample(n = 25) 

# display 
row2 
```

## Générer une proportion donné aléatoire d'un datasets

```python
# Sampling 25% of data ( randomly)
rows25 = data.sample(frac =.25) 
  
# checking if sample is 0.25 times data or not 
  
if (0.25*(len(data))== len(rows25)): 
    print( "Parfait") 
    print(len(data), len(rows25)) 
  
 #display 
rows25
```

```python
#Statistics from the sampling (25% of the employees datasets)


rows25.describe()
```

```python
# Statistics from the whole data sets

data.describe()
```

On voit trés bien que les estimations de la moyenne des `Salary` à partir de l'échantillon aléatoire qui represente 25% de la taille totale est proche de la moyenne de `Salary` de toute la population des employees du datasets.
Ceci estvalable pour le premier quartile , troisième quartile , std, min , max.
